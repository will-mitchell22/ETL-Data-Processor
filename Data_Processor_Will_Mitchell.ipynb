```python
import sqlite3
# This will create 5 tables within sqlite3
# The tables are listed below, with Merchandsie, Sections, Distributors, Transactions, and Items
def create_tables():
  try:
    conn = sqlite3.connect('inventory_management.db')
    cursor = conn.cursor()

    cursor.execute('''CREATE TABLE IF NOT EXISTS Merchandise(
                        merchandise_id INTEGER PRIMARY KEY,
                        merchandise_name TEXT NOT NULL,
                        price REAL NOT NULL,
                        quantity INTEGER NOT NULL,
                        category_id INTEGER,
                        FOREIGN KEY (category_id) REFERENCES Categories(category_id)
                  )''')

    cursor.execute('''CREATE TABLE IF NOT EXISTS Sections(
                        category_id INTEGER PRIMARY KEY,
                        category_name TEXT NOT NULL
                  )''')

    cursor.execute('''CREATE TABLE IF NOT EXISTS Distributors(
                        distributor_id INTEGER PRIMARY KEY,
                        distributor_name TEXT NOT NULL,
                        contact TEXT,
                        email TEXT,
                        phone TEXT
                  )''')

    cursor.execute('''CREATE TABLE IF NOT EXISTS Transactions(
                        transaction_id INTEGER PRIMARY KEY,
                        distributor_id INTEGER,
                        date_ordered DATE,
                        FOREIGN KEY (distributor_id) REFERENCES Distributors(distributor_id)
                  )''')

    cursor.execute('''CREATE TABLE IF NOT EXISTS Items(
                        item_id INTEGER PRIMARY KEY,
                        transaction_id INTEGER,
                        merchandise_id INTEGER,
                        quantity INTEGER,
                        price_total REAL,
                        FOREIGN KEY (transaction_id) REFERENCES Transactions(transaction_id),
                        FOREIGN KEY (merchandise_id) REFERENCES Merchandise(merchandise_id)
                  )''')

    conn.commit()
    conn.close()
    print("The tables were successfully created")
  except sqlite3.Error as e:
    print("Error Occured:", e)

create_tables()
```

```python
#This allows for us to work with the csv file and upload the data to Merchandise
import pandas as pd
def etl_csv(csv_file, Merchandise):
  try:
    df = pd.read_csv(csv_file)

    conn = sqlite3.connect('inventory_management.db')

    df.to_sql(Merchandise, conn, if_exists='replace', index=False)

    conn.commit()
    conn.close()
    print(f"Data from {csv_file} loaded into {Merchandise} table successfully.")
  except Exception as e:
    print("Error occured:", e)

uploaded = files.upload()

for filename in uploaded.keys():
  etl_csv(filename, 'Merchandise')
```
